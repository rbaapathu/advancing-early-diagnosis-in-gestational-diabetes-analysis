```python
df1.columns
```

```python
df1.drop(columns=['CCN'], inplace=True)
```

```python
# Convert 'Performance_Month' to datetime
df1['Performance_Month'] = pd.to_datetime(df1['Performance_Month'], format='%B')

# Convert 'State' and 'City' to categorical types
df1['State'] = df1['State'].astype('category')
df1['City'] = df1['City'].astype('category')
```

```python
# Calculate infection rate as Total Covid Infections per Total Resident Weeks
df1['Infection_Rate'] = (df1['Total_Covid_Infections'] / df1['Total_Resident_Weeks']) * 1000
```
